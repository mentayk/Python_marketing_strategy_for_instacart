### Creating if-statement with used-defined function for price range

```python
# Create user-defined function to define price lables for products: (low, mid, high ranged products)
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# Create a column for the user-defined function to get the results after using new function
df['price_range'] = df.apply(price_label, axis=1)
```

```python
# Value counts of price range to see results
df['price_range'].value_counts(dropna=False)
```

### Creating if-statement with for-loop for 'the busiest day'

```python
# Create for-loop, run value counts of order_day_of_week column
ords_prods_merged['order_day_of_week'].value_counts(dropna=False)
```

### As for numbers from value count 0 is the busiest day, and 4 is the least busy. Saturday is day 0, Wednesday is day 4 (by the data set)

```python
# Create a new column 'busiest day' with 3 different values:
result = []
for value in ords_prods_merged['order_day_of_week']:
    if value==0:
        result.append('Busiest day')
    elif value==4:
        result.append('Least busy')
    else:
        result.append('Regularly busy')
```

```python
# Attach a new column in the ords_prods_merged df
ords_prods_merged['Busiest day']=result
```

```python
# Run value counts for the result
ords_prods_merged['Busiest day'].value_counts(dropna=False)
```

## 2. Change the busiest day column to busiest and slowest DAYS column

```python
# 2. Create a column for busiest (0 and 1) and slowest (3 and 4) DAYS instead of DAY
# Value counts of order of the week column
ords_prods_merged['order_day_of_week'].value_counts(dropna=False)
```

```python
# Create a new column result_2 using for-loop with new requirements
result_2 = []

for value in ords_prods_merged['order_day_of_week']:
  if value == 0 or value == 1:
    result_2.append('Busiest days')
  elif value == 4 or value == 3:
    result_2.append('Slowest days')
  else:
    result_2.append('Regularly busy')
```

```python
# Attach the new column to ords_prods_merged df
# Run value counts to see new results
ords_prods_merged['Busiest days']=result_2
ords_prods_merged['Busiest days'].value_counts(dropna=False)
```

## 4. Create a column in ords_prods_merged df for the busy period (time) of the day

```python
# 4. Get value counts first to see the order time column closer
df['order_time'].value_counts(dropna=False)
```

### After getting the list of order time in ascending order, I split 24 to 3 to get 3 groups:
'Most orders'-10, 11, 14, 15, 13, 12, 16, 9
'Average orders'- 17, 8, 18, 19, 20, 7, 21, 22
'Fewest orders'- 23, 6, 0, 1, 5, 2, 4, 3

```python
# Create a new column busiest_time using for-loop with  requirements
busiest_time = []

for value in ords_prods_merged['order_time']:
  if value in [10, 11, 14, 15, 13, 12, 16, 9]:
    busiest_time.append('Most orders')
  elif value in [17, 8, 18, 19, 20, 7, 21, 22]:
    busiest_time.append('Average orders')
  else:
    busiest_time.append('Fewest orders')
```

```python
# Attach the new column to ords_prods_merged df
ords_prods_merged['Busiest_period_of_day']= busiest_time
```

```python
# 5. Print the frequency for the new column
ords_prods_merged['Busiest_period_of_day'].value_counts(dropna=False)
```
