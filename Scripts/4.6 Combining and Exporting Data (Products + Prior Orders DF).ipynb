 - Exporting dataframes in pkl format
 

```python
# Import libraries
import pandas as pd
import numpy as np
import os
```

```python
# Path to the original data
path= r"/Users/mentaykoshzhanova/Desktop/Instacart Basket Analysis"
path
```

```python
# 3. Import dataframes
df_ords_prods_combined= pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'orders_products_combined.pkl'))
df_prods= pd.read_csv(os.path.join(path, '02 Data', 'Prepared Data', 'products_clean.csv'), index_col = False)
df_ords= pd.read_csv(os.path.join(path, '02 Data', 'Prepared Data', 'orders_wrangled.csv'), index_col=False)
```

```python
# 6. Merging orders_products_combined df with products df with flag on to see if there is a full match
df_merged_large=df_ords_prods_combined.merge(df_prods, on='product_id', indicator = 'exists')
```
